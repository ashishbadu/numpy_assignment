```python
#Reversing 1D array

#using slicing
lis = np.array([1,2,3,4,5])
print(lis[::-1])

#using flip()
lis = np.array([1,2,3,4,5])
print(np.flip(lis))

```

```python
#Reversing 2D array

#using flip() function
lis = np.array([[1,2,3],[4,5,6],[6,7,8], [9,10,11]])
print(lis)
#reversed List along row
print("\nReversing 2D array along row:")
flip_lis = np.flip(lis , axis = 0)
print(flip_lis)

#reversed List along column side
print("\nReversing 2D array along column:")
flip_lis = np.flip(lis , axis = 1)
print(flip_lis)
#using
```

10. Explain the concepts of vectorization and broadcasting in NumPy. How do they contribute to efficient array operations?

ans. Vectorization is the process of performing operations on entire arrays or large sections of arrays without using explicit loops. This approach leverages NumPy's internal C and Fortran routines, making computations much faster and more efficient than equivalent Python loops.

```python
#with out vectorization
arr = np.array([1,2,3,4,5])
result = []
for x in arr:
  result.append(x ** 3)

print(f"without vectorization: {result} \n")

#with vectorization

result = arr ** 3
print(f"with vectorization: {result} \n")
```

#Broadcating
Broadcasting is a powerful mechanism that allows NumPy to work with arrays of different shapes when performing arithmetic operations. Broadcasting involves automatically expanding the dimensions of smaller arrays to match the dimensions of larger arrays.

```python
arr1 = np.array([1,2,3])
arr2 = np.array([[1],[2],[3]])
result = arr1 + arr2
print(result)
```

```python
arr1 = np.array([[1,2,3], [4,5,6]])
arr2 = np.array([1,2,3])
result = arr1 + arr2
print(result)
```

9. Generate a NumPy array of 100 random integers between 0 and 1000. Find and display all prime numbers in
this array.


```python
num = np.random.randint(0,1000,100)
print(num)
print('\n')

# displaying all prime numbers
for n in num:
    if n <= 1:
        print(n, "is not a prime number")
    else:
        is_prime = True

        for i in range(2, n+1):
            if (n % i) == 0:
                is_prime = False
                break
        if is_prime:
            print(n, "is a prime number")
        else:
            print(n, "is not a prime number")
```

10. Create a NumPy array representing daily temperatures for a month. Calculate and display the weekly averages.

```python
import numpy as np

# Create a NumPy array representing daily temperatures for a month
daily_temperatures = np.random.uniform(20, 30, 30)  # 30 days in a month

print("Daily Temperatures:")
print(daily_temperatures)

# Calculate the weekly averages
weekly_temperatures = []
for temp in range(0, 30, 7):
    weekly_average = np.mean(daily_temperatures[temp:temp+7])
    weekly_temperatures.append(weekly_average)

print("\nWeekly Averages:")
print(weekly_temperatures)
```
